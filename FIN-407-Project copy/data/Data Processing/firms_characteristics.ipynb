Chosen predictors:
- oibdpq: core profitability
- capxy: investment in equipment (sensitive to climate)
- invtq: can be sensitive to crop cycles or supply shocks
- actq + ancq = Total Assets: size indicator
- ltq / (actq + ancq) = Debt to Assets: leverage, risk from financial instability
- actq / lctq = Current Ratio: Liquidity, short-term health
- niq / (actq + ancq) = Return on Assets: general profitability
- (revtq - cogsq) / revtq = Gross Margin: core operational efficiency
- ∆revt = Sales Growth: reflects demand and production trends

```python
# Create predictors from firm characteristics
firm_characteristics = firm_characteristics.copy()

firm_characteristics['Total Assets'] = firm_characteristics['actq'] + firm_characteristics['ancq']
firm_characteristics['Debt-to-Assets'] = firm_characteristics['ltq'] / firm_characteristics['Total Assets']
firm_characteristics['Current Ratio'] = firm_characteristics['actq'] / firm_characteristics['lctq']
firm_characteristics['ROA'] = firm_characteristics['niq'] / firm_characteristics['Total Assets']
firm_characteristics['Gross Margin'] = (firm_characteristics['revtq'] - firm_characteristics['cogsq']) / firm_characteristics['revtq']
firm_characteristics['Sales Growth'] = firm_characteristics['revtq'].pct_change(fill_method=None)

# Drop no longer useful columns
firm_characteristics = firm_characteristics.drop(columns=['tic', 'conm', 'actq', 'ancq', 'ltq', 'lctq', 'niq', 'revtq', 'cogsq']).reset_index(drop=True)

# Rename the columns for consistence
columns_names = {'cusip':'CUSIP', 'datadate':'Date', 'oibdpq':'Operating Income', 'capxy':'CAPEX', 'invtq':'Inventories'}
firm_characteristics = firm_characteristics.rename(columns=columns_names)

# Format all dates to the end of month
firm_characteristics['Date'] = pd.to_datetime(firm_characteristics['Date'])
firm_characteristics['Date'] = firm_characteristics['Date'] + pd.offsets.MonthEnd(0)

display(firm_characteristics)
```
